```python
import os
import pandas as pd
```

```python
df = pd.read_excel('Legislative Districts by County and Town(1).xls', sheet_name='Senate', header=0)
df.info()
df.head()
```

```python
df = df[['Senate District', 'County']]
df.info()
df.head()
```

```python
df = df.drop_duplicates()
df.info()
df.head()
```

```python
df = df.reset_index(drop=True)
df.info()
df.head()
```

```python
grp = df.groupby(by=df['Senate District'].str.contains(',', na=False, regex=False))
grp
```

```python
grp.groups
```

```python
f = grp.get_group(False)
t = grp.get_group(True)
```

```python
f.info()
f.head()
```

```python
t.info()
t.head()
```

```python
x = t['Senate District'].str.strip()
y = x.str.split(',', expand=True)
y.info()
y.head()
```

```python
z = t[['County']].join(y, how='left')
z.info()
z.head()
```

```python
m = z.melt(id_vars=['County'])
m.info()
m.head()
```

```python
fin = m.drop(['variable'], axis=1)
fin.info()
fin.head()
```

```python
fin = fin.drop_duplicates()
fin.info()
fin.head()
```

```python
fin['Senate District'] = fin['value']
fin.info()
fin.head()
```

```python
fin2 = fin[['Senate District', 'County']]
fin2.info()
fin2.head()
```

```python
fin2 = fin2.dropna()
fin2.info()
fin2.head()
```

```python
df2 = pd.concat([f,fin2])
df2.info()
df2.head()
```
